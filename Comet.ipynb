```python
from os import listdir
from os.path import isfile, join, splitext
dataDir='/home/notebook/data/HUNT/'
filePat='comet-human-nextprot.txt'
allFiles = [f for f in listdir(dataDir) if f.endswith(filePat)]
processFiles=allFiles
#processFiles=[allFiles[1210]]
#processFiles=['121122_bsatest.mzML']
#processFiles=['/home/notebook/data/HUNT/141018_serum_508_tech2.raw.centroidcomet-human-nextprot.txt']
#processFiles.append(allFiles[1])
#processFiles.append(allFiles[2])
print(processFiles)
len(processFiles)
```

```python
processFiles=allFiles[:]
df=pd.DataFrame()
#for f in allFiles:
for f in processFiles:
    peptideHits=pd.read_csv(join(dataDir, f),sep='\t',skiprows=1,index_col=0)
    fnameExt = f.split('.')
    print(fnameExt[0])
    peptideHits['Name']=fnameExt[0]
    df=pd.concat([df,peptideHits],sort=False)
print(df.head())    
```

```python
dfDPcnt=df[df.delta_cn>1]
dfDPcnt=dfDPcnt[dfDPcnt.scan==1]
dfDPcnt=dfDPcnt['Name'].value_counts()
dfDPcnt.to_csv('count.csv', header=True)
print(dfDPcnt)
#dfDPcnt.to_csv((join(dataDir,'count.csv'))
```
