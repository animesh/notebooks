```python
import pandas as pd
data = pd.read_table("Jimita/txt48/proteinGroups.txt",sep='\t', index_col=0, header=0, lineterminator='\n')
data.head()
```

```python
data.shape
```

```python
dataLFQ=data.filter(like='LFQ', axis=1)
dataLFQ.shape
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
dataLFQlog2=np.log2(dataLFQ+1)
dataLFQlog2.hist()
```

```python
label = pd.read_table("Jimita/txt48/Groups.txt",sep='\t', index_col=0, header=0, lineterminator='\n')
label
```

```python
prot=dataLFQlog2.rename(columns=lambda x: x.replace('LFQ intensity ','')).transpose()
#prot=pd.merge(prot,label,left_on=['Protein IDs'],right_on=['Name'],how='outer')
prot=pd.merge(prot,label,left_index=True, right_index=True)
#prot=pd.concat([label,prot])
#prot.index
#label.index
prot.head(2)
```

```python
prot[prot>0].groupby(['Category']).median().S4R1W8.hist()
#prot.groupby(['Category']).sum(min_count=1)
```

```python
%matplotlib inline
from ipywidgets import interactive
def f(m):
    prot[prot>m].groupby(['Category']).median().S4R1W8.hist()
interactive_plot = interactive(f, m=(0.0, 30.0))
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot
```
